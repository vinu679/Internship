## 1. Print the count,min,max and standard deviation

```python
matches.describe()
```

```python
def count(matches):
    id=season=dl_applied=win_by_runs=win_by_wickets=list(matches.shape)[0]
    count=0
    l=list(matches.columns)
    for i in matches.values:
        if i[l.index('umpire3')]>0:
            #print(i[l.index('umpire3')])
            count+=1
    return id,season,dl_applied,win_by_runs,win_by_wickets,count        
count(matches)
```

```python
def mean(df):
    k=['id','season','dl_applied','win_by_runs','win_by_wickets']        #list(matches.columns)
    l=[]
    for i in k:
        l.append(sum(list(matches.loc[:,i].values))/len(matches.loc[:,i].values))
    return l
mean(matches)
```

```python
import math
def sd(df):
    l=mean(df)
    k=['id','season','dl_applied','win_by_runs','win_by_wickets']
    r=[]
    for i in k:
        s=0
        for j in list(df.loc[:,i].values):
            s+=pow((j-l[k.index(i)]),2)
        r.append(math.sqrt(s/(len(df.loc[:,i].values)-1)))
    return r
sd(matches)
```

```python
def minimum(df):
    k=['id','season','dl_applied','win_by_runs','win_by_wickets']
    r=[]
    for i in k:
        r.append(df[i].min())
        #r.append(min(df.loc[:,i].values))
    return r
minimum(matches)
```

```python
def maximum(df):
    k=['id','season','dl_applied','win_by_runs','win_by_wickets']
    r=[]
    for i in k:
        r.append(df[i].max())
        #r.append(max(df.loc[:,i].values))
    return r
maximum(matches)
```

## 4.Which team had won by maximum runs...?

```python
def TeamWonByMaxRuns(df):
    k=df['win_by_runs'].max()
    index=list(df.loc[:,'win_by_runs']).index(k)
    return df['winner'][index]
TeamWonByMaxRuns(matches)
```
